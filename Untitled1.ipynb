**UNEMPLOYMENT ANALYSIS WITH PYTHON :**

Unemployment is measured by the unemployment rate which is the number of people who are un employed as a percentage of the total labour force.We have seen a sharp increase in the employment rate during covid-19.

First, import the required libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

load the data:

```python
unemp = pd.read_csv('/content/Unemployment in India (1).csv')
```

group the data:

```python
grouped_unemp = unemp.groupby(['Region', 'Area'])
```

visualizing the data :

```python
plt.figure(figsize=(10,10))
plt.title("unemployment in india")
sns.barplot(x=' Estimated Unemployment Rate (%)',y= "Region", data=unemp)
plt.show()
```

The highest unemployment in tripura and lowest meghalaya.

```python
plt.figure(figsize=(10,10))
plt.title("unemployment in india")
sns.barplot(x=' Estimated Employed',y= "Region", data=unemp,)
plt.show()
```

The highest no.of people employed in uttarpradesh and lowest in sikkim

```python
plt.figure(figsize=(10,10))
plt.title("unemployment in india")
sns.barplot(x=' Estimated Labour Participation Rate (%)',y= "Region", data=unemp)
plt.show()
```

Estimated Labour Participation high in tripura and low in kerala.
